```python
df["dense_8"] = df["dense_8"].apply(lambda x:0 if x == -1 else x)
```

```python
df["label"].value_counts()
```

```python
import numpy as np
for dense in [f"dense_{x}" for x in range(1, 14)]:
    df[dense].fillna(int(df[dense].mean()), inplace=True)
    df[dense] = df[dense].apply(lambda x:np.log(x + 1))
```

```python
for sparse in [f"sparse_{x}" for x in range(1, 27)]:
    df[sparse].fillna("[UNK]", inplace=True)
```

```python
meta = df.describe(include='all').to_dict()
```

```python
for sparse in [f"sparse_{x}" for x in range(1, 27)]:
    meta[sparse]["vocab"] = df[sparse].unique().tolist()
    if "[UNK]" in meta[sparse]["vocab"]:
        meta[sparse]["vocab"].remove("[UNK]")
    meta[sparse]["vocab"].insert(0, "[UNK]")
```

```python
meta
```

```python
import json
class NpEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, np.integer):
            return int(obj)
        if isinstance(obj, np.floating):
            return float(obj)
        if isinstance(obj, np.ndarray):
            return obj.tolist()
        if isinstance(obj, bytes):
            return obj.decode("utf-8")
        return super(NpEncoder, self).default(obj)


with open(f"/data/day_{day}_meta.json", "w+") as fp:
    json.dump(meta, fp, cls=NpEncoder)
```
